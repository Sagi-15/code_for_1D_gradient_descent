# **Importing all libraries**

```python
import numpy as np
```

# **Loaded data and saw its type**

```python
data=np.loadtxt("data.csv",delimiter=",")
print(data) #so data is 2D numpy array
print("\n  Shape of data(rows,columns) : ",data.shape)
```

# **Input aur Output ko separate kar liya**

```python
input=[]
output=[]
for i in data:
  input.append(i[0])
  output.append(i[1])
x=np.array(input)
y=np.array(output)
print("Input : ",x)
print("Output : ",y)
```

# **Function step_gradient**

```python
def step_gradient(x,y,learning_rate,m,c):
  N=len(x) #so N mei total number of datapoints aaja
  m=m-(-2/N)*(learning_rate)*(((y-(m*x)-c)*(x)).sum())
  c=c-(-2/N)*(learning_rate)*((y-(m*x)-c).sum())
  return m,c
```

# **Function gradient_descent**

```python
def gradient_descent(x,y,learning_rate,number_of_iterations):
  m=0
  c=0
  #took initial m and c values yaha se start karke har iteration mei update karenge isko
  for i in range(number_of_iterations):
    m,c=step_gradient(x,y,learning_rate,m,c)
    print(i," ",m," ",c) #har step mei print m aur c
  return m,c
```

# **Function run() or main() ke similar**

```python
def run():
  #learning_rate=0.001 yeh karke can take feel of overfitting one positive one negative aur aise hote hi jaa bdhte ja
  learning_rate=0.000001
  number_of_iterations=100
  m,c=gradient_descent(x,y,learning_rate,number_of_iterations)
  print(m," ",c)
run()
```
